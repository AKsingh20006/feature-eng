```python
df[df['cgpa'_zscore] >3] | df[df['cgpa_zscore']<-3]
```

```python
#trimming
new_df = df[(df['cgpa_zscore']>3) &
(df['cgpa_zscore']<-3)]
new_df
```

```python
upper_limit = df['cgpa_zscore'].mean() + 3*df['cgpa_zscore'].std()
lower_limit = df['cgpa_zscore'].mean() - 3*df['cgpa_zscore'].std()
print(upper_limit)
print(lower_limit)
```

```python
df['cgpa'] = np.where(
    df['cgpa']>upper_limit,
    upper_limit,
    np.where(
        df['cgpa']<lower_limit,
        lower_limit,
        df['cgpa']
    )
)
print(df['cgpa'])
```
